```python
import pickle
import random

class TextGenerator(object):
    statistic_type_lookup = ['single_words', 
                             'double_words', 
                             'triple_words']
    probability_paragraph = 0.15
    
    def __init__(self, dump_filename):
        self.pre_last_word = False
        self.last_word = False
        
        pkl_file = open(dump_filename, 'r')
        self.words_statistic = pickle.load(pkl_file)
        pkl_file.close()

    def __iter__(self):
        return self
    
    def get_statistic_type(self):
        statistic_type = 0
                
        if self.pre_last_word:
            statistic_type = 2
        elif self.last_word:
            statistic_type = 1
            
        return statistic_type
    
    def get_random_word(self, statistic_type):
        new_word = False
        selection_words = self.words_statistic[self.statistic_type_lookup[statistic_type]]

        if statistic_type == 0:
            selection_words = selection_words.items()
        elif statistic_type == 1:
            selection_words = selection_words[self.last_word] if self.last_word in selection_words else []
        elif statistic_type == 2:
            selection_words = selection_words[(self.pre_last_word, self.last_word)] if (self.pre_last_word, self.last_word) in selection_words else []
               
        if len(selection_words) > 0:
            sum_selection_words = sum([value for word, value in selection_words])  
            random_value = random.randint(1, sum_selection_words)
            
            current_sum = 0            
            for key, value in selection_words:
                current_sum += value
                if current_sum >= random_value:
                    new_word = key
                    break

        return new_word
    
    def get_next_word(self):
        statistic_type = self.get_statistic_type()
        new_word = False
               
        while not new_word:
            new_word = self.get_random_word(statistic_type)
            statistic_type -= 1
            
        return new_word  
    
    def new_line_random(self):        
        if self.last_word == '.':
            if random.random() <= self.probability_paragraph:
                return '\n'
            
        return ''

    def next(self):
        new_word = self.get_next_word() 
                   
        if not self.last_word or self.last_word == '.':
            new_word = new_word.capitalize()
            
        if new_word != '.' and self.last_word:
            new_word = ' ' + new_word            
            
        self.pre_last_word = self.last_word
        self.last_word = new_word.lower().strip()  

        return new_word + self.new_line_random()
        
if __name__ == '__main__':
    text_generator = TextGenerator('word_statistic.pkl')
    text_iter = iter(text_generator)
    
    text = ""

    for i in xrange(10000):
        text += next(text_iter)
   
    print text
```
